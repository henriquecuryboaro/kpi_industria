```python
#imporar módulos apropriados para manipulação e anaĺise exploratória de dados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
def load_data(path_df):
    return pd.read_csv(path_df)

#importar base de dados
COMPLETE_PATH_EXPANDED = Path('__file__').resolve().parent / 'chemical_industry_kpis_complete_expanded.csv'
data = load_data(COMPLETE_PATH_EXPANDED)
```

```python
#Separar variáveis por tipo para aprofundamento em análises
def SeparaTipo(data):
    num_cols = data.select_dtypes(include = ['int64', 'float64']).columns
    cat_cols = data.select_dtypes(exclude = ['int64', 'float64']).columns
    var_numericas, var_categoricas = [column for column in num_cols], [column for column in cat_cols]
    return num_cols, cat_cols, var_numericas, var_categoricas

num_cols, cat_cols, var_numericas, var_categoricas = SeparaTipo(data)
```

```python
#Importação da biblioteca plotly.express para visualização de correlaçõaes
import plotly.express as px
```

```python
# correlações entre variáveis numéricas
num_data=data[var_numericas]
fig = px.imshow(num_data.corr(),text_auto=True)
fig.show()
# print(num_data.corr())
```

```python
import seaborn as sns

plt.figure(figsize=(12, 50))
for i, column in enumerate(num_cols):
    plt.subplot(20, 3, i+1)
    sns.boxplot(y=num_data[column])
    plt.title(f'Boxplot of {column}')
    plt.tight_layout()

plt.show()
```
